```python
import pandas as pd
import pathlib
```

```python
DATA_FOLDER = pathlib.Path(
    "/content/drive/MyDrive/gradebook_python/data"
)
```

```python
hw_exam_grades = pd.read_csv(
    DATA_FOLDER / "hw_exam_grades.csv",
    usecols=lambda x: "Submission" not in x,
    index_col="SID",
    converters={
        "SID": lambda x: x.lower()
    }
)
hw_exam_grades
```

```python
roster = pd.read_csv(
    DATA_FOLDER / "roster.csv",
    usecols=["Section", "Email Address", "NetID"],
    converters={
        "Email Address": lambda x: x.lower(),
        "NetID": str.lower
    },
    index_col="NetID"

)
roster
```

```python
pd.read_csv(DATA_FOLDER / "quiz_1_grades.csv")
```

```python
quiz_files = DATA_FOLDER.glob("quiz_*_grades.csv")
quiz_dataframe = pd.DataFrame()
for quiz_file in quiz_files:
    quiz_name = " ".join(quiz_file.name.split("_")[:2])
    quiz_df = pd.read_csv(
        quiz_file,
        index_col="Email",
        usecols=["Email", "Grade"],
        converters={
            "Grade": lambda x: float(x)
        }
    ).rename(columns={"Grade": quiz_name})
    quiz_dataframe = pd.concat([quiz_dataframe, quiz_df], axis=1)
quiz_dataframe
```

```python
final_data = pd.merge(
    roster,
    hw_exam_grades,
    left_index=True,
    right_index=True,
)
final_data
```

```python
final_data = pd.merge(
    final_data,
    quiz_dataframe,
    left_on="Email Address",
    right_index=True,
)
final_data
```

```python
final_data =  final_data.fillna(0)
```

```python
final_data.rename_axis("NetID")
```

# Calculate Homework Marks

```python
filter_homework_mks = final_data.filter(regex="Homework \d{1,2}$").sum(axis=1)
filter_homework_max_mks = final_data.filter(regex="Homework \d\d? -").sum(axis=1)
```

```python
final_data["Sum of Homework Scores"] = final_data.filter(regex="Homework \d{1,2}$").sum(axis=1)
final_data["Total Homework Scores"] = final_data.filter(regex="Homework \d\d? -").sum(axis=1)
final_data["Total Homework"] = filter_homework_mks / filter_homework_max_mks
final_data
```

```python
homework_max_points = final_data.filter(regex="Homework \d\d? -")
homework_scores = final_data.filter(regex="Homework \d{1,2}$")
hw_max_renamed = homework_max_points.set_axis(homework_scores.columns, axis=1)
avg_hw_scores = homework_scores / hw_max_renamed
final_data["Sum of Average Homework Score"] = avg_hw_scores.sum(axis=1)
```

```python
final_data["Average Homework"] = final_data["Sum of Average Homework Score"] / homework_scores.shape[1]
final_data
```
